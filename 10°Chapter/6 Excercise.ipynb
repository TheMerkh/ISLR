Consider the simple function R(β) = sin(β) + β/10.

Draw a graph of this function over the range β ∈ [−6, 6].

```python
import numpy as np
import matplotlib.pyplot as plt

def R(beta):
    return np.sin(beta) + beta / 10

def dR(beta):
    return np.cos(beta) + 1 / 10

beta_range = np.linspace(-6, 6, 400)
R_values = R(beta_range)

plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.plot(beta_range, R_values)
plt.grid(True)
```

What is the derivative of this function?

```python
dR_values = dR(beta_range)

plt.figure(figsize=(15, 5))
plt.subplot(122)
plt.plot(beta_range, dR_values)
plt.grid(True)
```

Given β 0 = 2.3, run gradient descent to find a local minimum
of R(β) using a learning rate of ρ = 0.1. Show each of β 0 , β 1 , . . .
in your plot, as well as the final answer.

```python
beta0 = 2.3
learning_rate = 0.1
num_iterations = 50

beta_history = [beta0]

for _ in range(num_iterations):
    gradient = dR(beta0)
    beta0 = beta0 - learning_rate * gradient
    beta_history.append(beta0)

plt.plot(beta_range, R_values)
plt.scatter(beta_history, [R(beta) for beta in beta_history], c="r")
plt.grid(True)
plt.show()
```
