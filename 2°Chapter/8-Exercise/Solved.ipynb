Q: Use the summary() function to produce a numerical summary of the variables in the data set. in this case i use describe()

Q: Use the plot() function to produce side-by-side boxplots of Outstate versus Private.

```python
#A:

pd.plotting.boxplot(df, column = ["Outstate"], color = "red", by = "Private", grid= False)
```

Q: Create a new qualitative variable, called Elite, by binning the Top10perc variable. We are going to divide universities into two groups based on whether or not the proportion of students coming from the top 10% of their high school classes exceeds 50%.

```python
#A:

df["Elite"] = ["No" if x <= 50 else "Yes" for x in df["Top10perc"]]
```

Q: Use the summary() function to see how many elite universities there are.

```python
#A:

df["Elite"].describe()
```

Q: Now use the plot() function to produce side-by-side boxplots of Outstate versus Elite.

```python
#A:

pd.plotting.boxplot(df, column =["Outstate"],by ="Elite", grid = False)
```
